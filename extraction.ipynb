```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data from Excel file
weeklyReportDict = pd.read_excel('data2024.xlsx', sheet_name=None)

# Initialize dictionary to hold sums
myDict = {}

# Process and plot Attendance Forms
for eachWeek, data in weeklyReportDict.items():
    dfOfEachW = pd.DataFrame(data)
    print('Week: ', eachWeek, 'Total: ', np.sum(dfOfEachW.iloc[8:10, 2]))
    myDict[eachWeek] = np.sum(dfOfEachW.iloc[8:10, 2])

# Reverse the order of weeks and totals for plotting
weeks = list(reversed(myDict.keys()))
totals = list(reversed(myDict.values()))

# Plot Attendance Forms data
plt.figure(figsize=(15, 7))
plt.plot(weeks, totals, marker='o', linestyle='-')
plt.xlabel('Weeks', fontsize=14)
plt.ylabel('Forms', fontsize=14)
plt.title('Total Attendance Forms Processed', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Adjust y-ticks
current_y_ticks = plt.gca().get_yticks()
plt.gca().set_yticks(current_y_ticks)
plt.gca().set_yticklabels([f'{int(y):,}' for y in current_y_ticks], fontsize=12)

plt.grid(True)
plt.tight_layout()
plt.show()

# Process and plot Curriculum Forms
myDict = {}
for eachWeek, data in weeklyReportDict.items():
    dfOfEachW = pd.DataFrame(data)
    print('Week: ', eachWeek, 'Total: ', np.sum(dfOfEachW.iloc[6:8, 2]))
    myDict[eachWeek] = np.sum(dfOfEachW.iloc[6:8, 2])

# Reverse the order of weeks and totals for plotting
weeks = list(reversed(myDict.keys()))
totals = list(reversed(myDict.values()))

# Plot Curriculum Forms data
plt.figure(figsize=(15, 7))
plt.plot(weeks, totals, marker='o', linestyle='-')
plt.xlabel('Weeks', fontsize=14)
plt.ylabel('Forms', fontsize=14)
plt.title('Total Curricula Forms Processed', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Adjust y-ticks
current_y_ticks = plt.gca().get_yticks()
plt.gca().set_yticks(current_y_ticks)
plt.gca().set_yticklabels([f'{int(y):,}' for y in current_y_ticks], fontsize=12)

plt.grid(True)
plt.tight_layout()
plt.show()
```
